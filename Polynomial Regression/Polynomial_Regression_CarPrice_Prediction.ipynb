```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
X = df[["enginesize"]]
y = df["price"]
```

```python
plt.figure()
plt.scatter(X, y)
plt.xlabel("Engine Size")
plt.ylabel("Car Prize")
plt.title("Relationship between Engine Size and Price")
plt.show()
```

### Insights : 

The plot shows a strong positive correlation between engine size and car price

As engine size increases, the car price generally increases

```python
linear_model= LinearRegression()
linear_model.fit(X,y)
```

```python

from sklearn.pipeline import Pipeline
poly_model = Pipeline([
    ("poly", PolynomialFeatures(degree=4)),
    ("linear", LinearRegression())
])
poly_model.fit(X, y)
```

```python
test_engine_size = pd.DataFrame({"enginesize": [200]})

linear_prediction = linear_model.predict(test_engine_size)[0]
poly_prediction = poly_model.predict(test_engine_size)[0]

print("Linear model prediction:", linear_prediction)
print("Polynomial model prediction:", poly_prediction)
```

```python
engine_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)

linear_curve = linear_model.predict(engine_range)
poly_curve = poly_model.predict(engine_range)
 
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(engine_range, linear_curve, label="Linear Model")
plt.plot(engine_range, poly_curve, label="Polynomial Model")
plt.xlabel("Engine Size")
plt.ylabel("Car Price")
plt.title("Linear vs Curved Model Comparison")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
# Polynomial transformation
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

# Train model
linear_model = LinearRegression()
linear_model.fit(X_poly, y)

# Prediction
y_pred = linear_model.predict(X_poly)
```

```python
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print("Polynomial Regression Metrics")
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("R2 Score:", r2)
```
